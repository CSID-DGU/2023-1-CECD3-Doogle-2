# 1. 한글 데이터 불러오기

```python
path = './sen_data/train/morpheme/'
import os
os.listdir(path)
```

```python
import collections
```

```python
# file path list generate

folder_file_path = collections.defaultdict(list)

for folder_name in os.listdir(path):
  folder_path = path + folder_name # '/content/drive/MyDrive/종설/data/sen_data/train/morpheme/06'
  folder_file_path[folder_name] = os.listdir(folder_path)

folder_file_path
```

```python
# file open
def ksl_df_gen(folder_num):
  ksl_sen_data = [] # 한국 수어 문장 담은 리스트
  file_paths = folder_file_path[folder_num]
  for file_path in tqdm(file_paths):
    file_path = path + folder_num + '/' + file_path
    with open(file_path, 'r') as file:
      json_data = json.load(file)
      ksl_sen = []  # 한 문장 안에 속하는 수어 단어들을 담을 리스트
      for ksl_dict in json_data['data']:  # json_data['data']는 딕셔너리 원소의 리스트 형태
      # ksl_dict은 한 딕셔너리 원소
        ksl_sen.append(ksl_dict['attributes'][0]['name'])

      ksl_sen_data.append(ksl_sen)

  # 데이터 프레임 생성
  # 현재 폴더 명 변수 생성
  folder_nums = np.repeat(int(folder_num),len(ksl_sen_data))
  ksl_sen_df = pd.DataFrame(folder_nums)
  ksl_sen_df.columns=['folder_nums']

  # 한국어 수어 단어 리스트 변수 생성
  ksl_sen_df['ksl_sen'] = list(ksl_sen_data)

  return ksl_sen_df
```

```python
# ksl_sen_df 생성
ksl_sen_df = ksl_df_gen(os.listdir(path)[0])  # os.listdir(path)[0] is '02'
ksl_sen_df
```

```python
folder_list = os.listdir(path)[2:]
folder_list
```

```python
ksl_sen_df
```

```python
# folder 돌면서
for folder_num in folder_list:
  ksl_sen_df_tmp = ksl_df_gen(folder_num)
  ksl_sen_df = pd.concat([ksl_sen_df, ksl_sen_df_tmp])
```

# 영어 수어 데이터

```python
ksl_sen_df = pd.read_csv('./ksl_sen_df.csv')
ksl_sen_df
```

```python
import googletrans

g_trans = googletrans.Translator()
```

```python
ksl_sen_df
```

```python
import time
```

```python
# 한국어 수어 -> google translator api 이용해 영어 수어로 번역
ksl_sen_trans = []
for i in tqdm(range(len(ksl_sen_df))):
  for sen in ksl_sen_df.loc[i,'ksl_sen']:
    ksl_sen_trans_tmp = []
    #print(word)
    for word in sen:
      word_trans_res = g_trans.translate(word, dest='en',src='ko')
      time.sleep(1)
      ksl_sen_trans_tmp.append(word_trans_res.text)
    ksl_sen_trans.append(ksl_sen_trans_tmp)
  #print(ksl_sen_trans)

ksl_sen_df['ksl_sen_trans'] = list(ksl_sen_trans)

ksl_sen_df
```

```python
ksl_sen_df.to_csv('/content/drive/MyDrive/종설/data/ksl_sen_df_trans.csv',index=False)
```
